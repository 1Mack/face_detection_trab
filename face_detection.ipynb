```
# Carregando as dependÃªncias
import numpy as np
import pandas as pd
import cv2 as cv 
from google.colab.patches import cv2_imshow
from skimage import io
from PIL import Image 
import matplotlib.pylab as plt
import scipy
```

```
# Atribuindo uma imagem
urls = ["https://amenteemaravilhosa.com.br/wp-content/uploads/2015/10/rostos-confianca.jpg"]

# Mostrando imagem escolhida
for url in urls:
  image = io.imread(url) 
  image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
  
cv2_imshow(image_2)
```

```
## Detectando faces colocando retangulo nelas

# tÃ©cnica pra detectar faces humanas
face_cascade = cv.CascadeClassifier('haarcascade_frontalface_default.xml') # arquivo para detectar face

# Eliminando falsos positivos, definindo parÃ¢metro de escala e tamanho mÃ­nimo do objeto
faces = face_cascade.detectMultiScale(image_2, scaleFactor = 1.04, minNeighbors = 7, minSize = (30,30), flags = cv.CASCADE_SCALE_IMAGE)

# Definindo cor ao retÃ¢ngulo e passando coordenadas:
for (x,y,w,h) in faces:
     cv.rectangle(image_2,(x,y),(x+w,y+h),(0,191,255),2)

cv2_imshow(image_2)
```

```
## Inserindo CÃ­rculo nas faces
for (x, y, w, h) in faces:
    coordinates = x + w // 2, y + h // 2 # coordenadas do ponto mÃ©dio
    radius = w // 2 # raio do cÃ­rculo
    cv.circle(image_2, coordinates, radius, (255, 0, 255), 3) # Definindo cor aos circulos e passando coordenadas

cv2_imshow(image_2)
```
